```python
def remove_noise(text):
    clean_text = re.sub(r'[\d!@#$%^&*()?><:"|;\'.\[\]]', "", str(text))
    clean_text = re.sub(r'[^A-Za-z]+', " ", clean_text)
    return clean_text.lower()

df["kyewords"] = df["kyewords"].apply(remove_noise)
```

```python
df["kyewords"][0]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

def remove_stopwords(text):
    return " ".join([i for i in text.split() if i not in stopwords.words("english")])

df["kyewords"] = df["kyewords"].apply(remove_stopwords)
```

```python
# !python -m spacy download en_core_web_sm
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def word_lem(text):
    doc = nlp(text)
    lemmatized_text = " ".join([i.lemma_ for i in doc])
    return lemmatized_text

df["kyewords"] = df["kyewords"].apply(word_lem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(df["kyewords"])
```

```python
x.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel
similarity = linear_kernel(x)
```

```python
similarity
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]]["title"])
```

```python
recommend("Spider-Man")
```

```python
import pickle

with open('similer_matrix.pkl', 'wb') as file:
    pickle.dump(similarity, file)
```
